```python
import operator

all = []
sort = []
with open("/content/drive/MyDrive/popular-names.txt") as f:
  for line in f:
    word = line.strip().replace("\t"," ")
    w = word.split(' ')
    all.append(w)
  sort = sorted(all, key=operator.itemgetter(2), reverse=True)
  print(sort)
```
